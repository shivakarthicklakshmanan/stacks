````python
def split_text_into_chunks(text, lines_per_chunk=10):
    # Split the text into lines
    lines = text.split('.')
    #print(lines)
    # Create chunks of specified size
    chunks = ['.'.join(lines[i:i + lines_per_chunk]) for i in range(0, len(lines), lines_per_chunk)]
    #print(text)
    return chunks
````

````python
import PyPDF2
import pandas as pd

# Path to the uploaded PDF file
pdf_path = 'Final_SE_Conferenc-Transcript.pdf'

# Function to extract text from specific pages of the PDF
def extract_text_from_pdf(pdf_path, start_page, end_page):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(start_page - 1, end_page):  # PyPDF2 is 0-indexed
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

# Extract text from pages 3 to 16
pdf_text = extract_text_from_pdf(pdf_path, 3, 16)

# Split the text based on speaker names (this may vary depending on the transcript format)
# Example: Assuming each speaker section starts with their name followed by their transcript
lines = pdf_text.split('\n')
speakers = []
texts = []
current_speaker = None
current_text = []
conv = dict()
counter = 0

for line in lines:
    if ':' in line:  # Simple heuristic to identify a speaker line
      #print(line)
      #print(line.split(':')[0])
      if line.split(':')[0] in ['Raju Barnawal','Rohan', 'Omkar','Moderator','Vivek Wahi','Sushil Choksey','Ashok Ajmera','M. V. Rao','Ashlesh Sonje','Mukul Dandige','Moderator']:
        speaker = line.split(':')[0]
        #print(speaker)
        #texts.append(line.strip().split(':')[1])
        text = line.split(':')[1]
        #print(text)
        conv[speaker] = str(counter)+','+text.strip()
        counter += 1


def split_number_text(text):
    number, message = text.split(",", 1)
    return int(number), message

# Create a list of tuples with the speaker, number, and message
sorted_data = [(speaker, *split_number_text(text)) for speaker, text in conv.items()]

# Sort the list by the number
sorted_data.sort(key=lambda x: x[1])

# Create a DataFrame from the sorted data
df = pd.DataFrame(sorted_data, columns=['Speaker', 'Sequence Number', 'Text'])
df.sort_values(by='Sequence Number', ascending = True, inplace=True)
df
````

````python
file_path = 'conversation.txt'

# Open the file in write mode
with open(file_path, 'w') as f:
    # Iterate through the DataFrame and write each row in the desired format
    for index, row in df.iterrows():
        f.write(f"{row['Speaker']}: {row['Text']}\n")

print(f"Data has been written to {file_path}")
````
